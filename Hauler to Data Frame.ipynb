```python
def dowToNum(val):
    lookup={"MON":0,"TUE":1,"WED":2,"THU":3,"FRI":4}
    if val != val: return -1
    return lookup[val]
```

```python
indata["PickupDay"].unique()
```

```python
indata["PickupDay"] = indata["PickupDay"].apply(dowToNum)
indata
```
